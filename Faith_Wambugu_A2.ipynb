```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
listings = pd.read_csv("Airbnb_NYC_2019.csv")
```

```python
listings
```

```python
#Create a histogram for all prices
plt.hist(listings["price"], bins = np.arange(0, 1100, 40))
plt.title("Distribution of Airbnb NYC listing prices in 2019")
plt.xlabel("Price(in US dollars)")
plt.ylabel("Count")
plt.show()
```

```python
#Create a scatter plot of price over no. of reviews
plt.scatter(x = listings["price"],y = listings["number_of_reviews"])
plt.title("Scatter plot for Number of reviews vs Price of Airbnb NYC Listings for 2019")
plt.xlabel("Price(in US dollars)")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
#If you want to zoom in on x axis
plt.scatter(x = listings["price"],y = listings["number_of_reviews"], s = 5)
plt.xlim(0, 1100)
plt.title("Scatter plot for Number of reviews vs Price of Airbnb NYC Listings for 2019")
plt.xlabel("Price(in US dollars)")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
#bar chart that displays the count of bnb listings 
# in each neighborhood group of NYC from listings df
sn.countplot(x = "neighbourhood_group", data = listings )
plt.title("Airbnb listings per neighbourhood group in 2019")
plt.show()
```

```python
#Barchart for ave. price of Airbnb listings in 
#each neighbourhood group of NYC from listings df
#black lines mark confidence intervals
sn.barplot(x = "neighbourhood_group", y = "price", data = listings)
plt.show()
```

```python
# ci =  false to remove the black lines
sn.barplot(x = "neighbourhood_group", y = "price", data = listings, ci = False)
plt.show()
```

```python
#Select a subset of manhattan neighbourhood group from listings df
listings.loc[["Manhattan" in neighbourhood_group for neighbourhood_group in listings["neighbourhood_group"]],:]
```

```python
#Give the subset a df
listings_subset = listings.loc[["Manhattan" in neighbourhood_group for neighbourhood_group in listings["neighbourhood_group"]],:]
```

```python
listings_subset
```

```python
#No. of rows and columns in the subset
listings_subset.shape
```

```python
#Countplot for all neighbourhoods of Manhattan
plt.figure(figsize=(18, 9))
sn.countplot(y = "neighbourhood",  data = listings_subset)
plt.title("Top 10 Manhattan neighbourhood bnbs in 2019")
plt.show()
```

```python
#neighbourhoods with ten highest counts
#listings_Manhattan10 = listings_subset.sort_values("neighbourhood", ascending = False)#.groupby('neighbourhood')
#listings_Manhattan10 = listings_subset['neighbourhood'].value_counts(ascending=False).reset_index()
#listings_Manhattan10 = listings_subset['neighbourhood'].value_counts()
#listings_Manhattan10 = listings_subset.groupby('neighbourhood').count()
listings_Manhattan10 = listings_subset.value_counts(subset='neighbourhood',sort=True,ascending=False)
listings_Manhattan10 
```

```python
#Order of 1st to 10th highest count
listings_Manhattan10.iloc[0:10]
```
